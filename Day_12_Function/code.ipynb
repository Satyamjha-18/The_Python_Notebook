### Positional Arguments

```python
def show(name,age):
    print('name is' , name)
    print('age is', age)

    
```

```python
show(21,"sanya")
```

### keyworded arguments

```python
def show(name,age):
    print('name is' , name)
    print('age is', age)

    
```

```python
show(name="sanya",age=22)
```

```python
show(age=22,name="sanya")
```

### default argument 

```python
def show(name='Raj'):
    print(name)
```

```python
show() # not pass anything  it will automatic take the default parameter
```

```python
show("Sanya") # replace the default value
```

```python
def total(numbers):
    x=0

    for n in numbers:
        x+=n

    return x
print(total([5,10,15]))    
```

```python
def large(a,b):
    if a >b:
        return a
    else:
        return b
print(large(12,25))
```

```python
def even_odd(num):

    if num % 2 == 0:
        return "Even"

    return "Odd"

print(even_odd(15))
```

```python
def largest(a, b):

    if a > b:
        return a

    return b

print(largest(20, 15))
```

```python
def total(*numbers):
    print(sum(numbers))

total(10,20,30,40)
```

```python
def details(**info):

    for key, value in info.items():
        print(key, ":", value)

details(name="Satyam", age=22, city="Patna")
```
