### List method


```python
fruits = ["Apple","Banana","Mango"]
```

```python
fruits
```

```python
# 1. insert()

fruits.insert(1,"Pineapple")

print(fruits)
```

```python
# 2. append()


fruits.append("Kiwi")

print(fruits)
```

```python
# 3. extend()


fruits.extend(["Cherry","Papaya"])

print(fruits)
```

```python
# 4. remove()

fruits.remove("Apple")

print(fruits)
```

```python
# 5. pop()


fruits.pop()

print(fruits)
```

```python
# 6. sort()

numbers = [50,10,40,20,30]

numbers.sort()

print(numbers)
```

```python
# 7. reverse()

numbers.reverse()

print(numbers)
```

```python
for item in fruits:
    print(item)
```

```python
square = [x*x for x in range(1,11)]

print(square)
```

```python
#  Find Maximum

numbers = [12,45,67,89,23]

print(max(numbers))
```

```python
# Sum of List


print(sum(numbers))
```

```python
# Copy List


copy_list = numbers.copy()

print(copy_list)
```

```python
#  Nested List


matrix = [[1,2],[3,4],[5,6]]

for row in matrix:
    print(row)
```

```python
numbers = []

n = int(input("How many elements? "))

for i in range(n):
    value = int(input("Enter Number: "))
    numbers.append(value)

print("List =", numbers)
print("Largest =", max(numbers))
print("Smallest =", min(numbers))
print("Sum =", sum(numbers))
```
