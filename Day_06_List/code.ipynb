### List comprehension : list comprehension provides a concise way of creating list.

* new list =[expression for item in iterable if condition== True]

* Advantage
  1. More time-efficient and space-efficient then   loop
  2. Require fewer lines of code

  3. transforms iterative statement into a formula

```python
# Add 1 to 10 number to a list
l =[]
for i in range(1,11):
    l.append(i)
print(l)
```

```python
l=[i for i in range(1,11)]
l
```
