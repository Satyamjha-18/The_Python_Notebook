```python
i = 1
while i<6:
 print(i)
 i = i+1
```

```python
#Write a while loop to print Hi 7 times on the screen.
i = 1
while i<=7:
 print(i, "Hi")
 i = i+1
```

```python
i = 1
while i<=7:
 print(i, "Hi")
 i += 1 # i = i+1
```

```python
#Print all the even numbers between 0 to 10 (both inclusive) using a while loop
i = 0
while i<= 10:
 print(i)
 i += 2
```

```python
i = 0
while i<= 10:
 if i%2 == 0:
  print(i)
 i += 1
```

```python
#Print all the odd numbers between 0 to 10.
i = 1
while i<= 9:
  print(i)
  i += 2
```

```python
i = 0 
while i <10:
    if i %2!=0:
     print(i)
    i+=1
```

```python
# Print the sum of the number between 1 to 10 (both inclusive)
sum = 0
for i in range(1,11):
 sum += i
 print(sum)
```
