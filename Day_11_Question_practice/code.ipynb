```python
# given a string write a code to convert it to lowercase. U cannot use lower

s =" Raj" 

new_s=""
for ch in s:
    if ch<='Z' and ch >='A':
     new_s += chr(ord(ch)+32)

    else:
       new_s+=ch

print(new_s)

```
