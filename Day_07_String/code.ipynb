```python
# ==========================================
# 10. String Methods

text = " python programming "

print(text.upper())

print(text.lower())

print(text.title())

print(text.capitalize())

print(text.strip())

print(text.replace("python", "Java"))
```

```python
# 11. Split and Join

sentence = "Python is Easy"

words = sentence.split()

print(words)

print("-".join(words))
```
