```python
# 12. Simple Calculator


a = float(input("First Number: "))
b = float(input("Second Number: "))
op = input("Operator (+,-,*,/): ")

if op == "+":
    print(a + b)
elif op == "-":
    print(a - b)
elif op == "*":
    print(a * b)
elif op == "/":
    if b != 0:
        print(a / b)
    else:
        print("Division by zero is not allowed")
else:
    print("Invalid Operator")
```
