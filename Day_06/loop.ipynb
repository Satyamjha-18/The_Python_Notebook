```python
# print all the int from 1 to 5

for i in range(1,6):
    print(i)
```

```python
i =1
while i<6:
    print(i)
    i=i+1
```

```python
i =1 
while i <8 :
    print(i,"hii")
    i = i +1
```

```python
i =1 
while i <8 :
    print(i,"hii")
    i += 1
```

```python
i = 7
while i>=1:
    print(i,"hi")
    i-=1
```
